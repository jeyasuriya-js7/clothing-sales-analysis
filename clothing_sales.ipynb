```python
clothing["Year"] = clothing["saleDate"].dt.year
clothing["Month"] = clothing["saleDate"].dt.month_name()
clothing["Month Number"] = clothing["saleDate"].dt.month
print(clothing)
```

```python
profit_by_salesperson=clothing.groupby('salespersonName')['profit'].sum().sort_values(ascending=False).head(10)
print(profit_by_salesperson)
```

```python
clothing['profit']=clothing['totalAmount']-clothing['totalCost']
print(clothing)
```

```python
clothing['profit']=clothing['profit'].round(2)
clothing.to_csv("sales_data_updated.csv",index=False)
```

```python
total_sales=clothing['totalAmount'].sum()
print("total sales:",total_sales)
total_profit=clothing['profit'].sum()
print("total profit:",total_profit)
```

```python
profit_perc=clothing['profit']/clothing['totalAmount']*100
print(profit_perc)
```

```python
import matplotlib.pyplot as plt
import numpy as np
most_profit_by_location=(
    clothing.groupby('location')['totalAmount'].sum().sort_values(ascending=False).head(5) 
) 
plt.figure(figsize=(6,5))
plt.bar( 
    most_profit_by_location.index,
    most_profit_by_location.values
) 
plt.title('top sales by location')
plt.xlabel('totalAmount') 
plt.ylabel('location')
plt.show()
```
